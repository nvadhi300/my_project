```python
filtered_bnf_df = df_bnf[df_bnf['BNF Name'].str.contains('levonorgestrel', case=False, na=False) |
                         df_bnf['DM+D: Product Description'].str.contains('levonorgestrel', case=False, na=False)]
filtered_bnf_df
```

```python
# Custom function to find matches
def find_matches(row, df2):
    matches = df_bnf[df_bnf['BNF Name'].str.contains(row['Drug/device name'], case=False, na=False)]
    if not matches.empty:
        return pd.concat([pd.DataFrame([row]*len(matches)).reset_index(drop=True), matches.reset_index(drop=True)], axis=1)
    else:
        return pd.DataFrame()

# Apply the function to each row in df1
matched_df = pd.concat([find_matches(row, df_bnf) for index, row in filtered_pca_df.iterrows()], ignore_index=True)

#print(matched_df)

# Ensure 'SNOMED Code' is treated as a string
matched_df['SNOMED Code'] = matched_df['SNOMED Code'].astype(str)

# Selecting specific columns from the matched DataFrame
pca_snomed_Data = matched_df[['BNF chemical name', 'Drug/device name','BNF Name', 'BNF Code','SNOMED Code', 'Drug formulation', 'Drug strength']]

pca_snomed_Data

pca_snomed_Data.to_csv('pca_snomed_data.csv', index=False)
```

```python
# Function to filter df_bnf based on partial string matching
def find_matches_optimized(df1, df2):
    # Filter df2 based on whether 'BNF Name' contains any 'Drug/device name' from df1
    mask = df2['BNF Name'].apply(lambda x: any(name in x for name in df1['Drug/device name'].values))
    filtered_df2 = df2[mask]

    # Perform a merge where 'BNF Name' contains 'Drug/device name'
    merged_df = pd.merge(df1, filtered_df2, left_on='Drug/device name', right_on='BNF Name', how='inner')

    return merged_df

# Apply the optimized function
matched_df = find_matches_optimized(filtered_pca_df, df_bnf)

# Ensure 'SNOMED Code' is treated as a string
matched_df['SNOMED Code'] = matched_df['SNOMED Code'].astype(str)

# Selecting specific columns from the matched DataFrame
pca_snomed_Data = matched_df[['BNF chemical name', 'Drug/device name', 'BNF Name', 'BNF Code', 'SNOMED Code', 'Drug formulation', 'Drug strength']]

# Saving the final dataframe to a CSV file
pca_snomed_Data.to_csv('pca_snomed_data.csv', index=False)
```
